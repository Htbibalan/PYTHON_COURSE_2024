```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
figfolder= r"C:\Users\hta031\Github\PYTHON_COURSE_2024\plots\\"
```

```python
file_path = r"C:\Users\hta031\Github\PYTHON_COURSE_2024\source\Pearson-1.txt"
data = np.loadtxt(file_path, skiprows=1)
# to show all the rows of data
np.set_printoptions(threshold=np.inf)
print(data)
```

```python
with open(file_path) as file:
    file_contents = file.read()
    print(file_contents)

```

```python
fathers= data[:,0]*2.4
sons= data[:,1]*2.4
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=fathers,y=sons, color= "orange")
plt.title("Scatter plot of Father and Son")
plt.show()

#plt.savefig(figfolder + "scatter.jpeg")


f, ax = plt.subplots()
ax.hist(fathers, color="gold", bins=20)
ax.hist(sons, color="gray", alpha= 0.3, bins=20)
ax.set_title("Histogram of fathers and sons")
ax.set_ylabel("frequency")

# plot meadian 
ax.axvline(np.median(fathers))
ax.axvline(np.median(sons), color= "gray")
f.savefig(figfolder + "histo_heights.jpeg")



plt.figure(figsize=(10, 6))
sns.regplot(x=fathers,y=sons, marker = "+",line_kws={"color": "purple"})
plt.title("Relationship between Fathers and Sons")
plt.show()


f, ax=plt.subplots()
ax.scatter(fathers,sons, s=30, alpha = 0.5)
plt.show()



# correlation =fathers.corr(sons)
# print("The correlation between X and Y is", correlation)
# # #calculate the covariance
# # covariance = data["Father"].cov(data["Son"])
# print("The covariance between X and Y is", covariance)
```

```python
f, ax = plt.subplots()
ax.scatter(fathers, sons, s=20, alpha=0.5)

correlation_coefficient = np.corrcoef(fathers, sons)[0, 1]


fit = np.polyfit(fathers, sons, deg=1)
ax.plot(fathers, fit[0] * np.array(fathers) + fit[1], color='red', linewidth=2)


ax.text(0.1, 0.9, f'Correlation Coefficient: {correlation_coefficient:.2f}', transform=ax.transAxes, fontsize=10, color='red')


plt.show()
f.savefig(figfolder+ "scatter_cor.jpeg")
```
