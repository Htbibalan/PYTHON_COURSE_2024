```python
def tweak_df(df):
    return (df
            .rename(columns=str.lower)
            .rename(columns={"total_awards":"awards"})
            .replace("White/European", "white")
            .astype({"race":"int"}) # this converts the type of the value from string to int that might change the memory dedicated to each value which comes in handy when we have a large dataset
            
            )   
```

```python
tweak_df(df_artists)
```
